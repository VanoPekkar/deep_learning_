```python
def copy_files(filenames, src_dir, dest_dir):
    for filename in filenames:
        shutil.copy(os.path.join(src_dir, filename),
                    os.path.join(dest_dir, filename))

def prepare_data(src_path, train_dir, val_dir, test_size=0.2, seed=42):
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)
    images = [f for f in os.listdir(src_path) if f.endswith('.jpg')]
    # Разделение на обучающую и валидационную выборку
    train_imgs, val_imgs = train_test_split(images, test_size=test_size, random_state=seed)
    copy_files(train_imgs + [img.replace('.jpg', '.txt') for img in train_imgs], src_path, train_dir)
    copy_files(val_imgs + [img.replace('.jpg', '.txt') for img in val_imgs], src_path, val_dir)

# Пути до папок
data_src = "/kaggle/input/traffic-signs-dataset-in-yolo-format/ts/ts"
train_dst = "/kaggle/working/dataset/train"
val_dst = "/kaggle/working/dataset/val"

# Подготовка датасета для yolo
prepare_data(data_src, train_dst, val_dst)
```

```python
# Файл с названием классов
yaml_content = """
train: /kaggle/working/dataset/images/train
val: /kaggle/working/dataset/images/val

nc: 4
names: ['speed limit', 'yield', 'mandatory', 'other']
"""

with open('/kaggle/working/dataset/dataset.yaml', 'w') as file:
    file.write(yaml_content)
```
